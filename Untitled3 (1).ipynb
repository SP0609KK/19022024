```python
from PIL import Image, ImageDraw
import json

# Load JSON data from file
with open('result2.json', 'r') as f:
    data = json.load(f)

# Load the image
image_path = '/content/pic1.png'
image = Image.open(image_path)
draw = ImageDraw.Draw(image)

# Define the category colors dictionary
# Define the category colors dictionary
category_colors = {
    0: "red",
    1: "blue",
    2: "green",
    3: "yellow",
    4: "orange",
    5: "purple",
    6: "pink",
    7: "cyan",
    8: "magenta",
    9: "lime",
    10: "teal",
    11: "maroon",
    12: "olive",
    13: "navy",
    14: "brown",
    15: "gray",
    16: "black",
    # Add more category IDs and colors as needed
}

# Extract category names from JSON data
categories = {category['id']: category['name'] for category in data['categories']}

# Iterate through annotations
annotations = data['annotations']
for annotation in annotations:
    category_id = annotation['category_id']

    # Check if the category ID is in the color dictionary and category names
    if category_id in category_colors and category_id in categories:
        bbox = annotation['bbox']

        # Extract bounding box coordinates
        x1, y1, width, height = bbox
        x2 = x1 + width
        y2 = y1 + height

        # Draw rectangle
        draw.rectangle([x1, y1, x2, y2], outline=category_colors[category_id], width=2)

        # Get the category name using the category ID
        category_name = categories[category_id]

        # Label the box with category name
        draw.text((x1, y1), category_name, fill=category_colors[category_id])
    else:
        print(f"Category ID {category_id} not found in category_colors dictionary or categories list.")

# Display the image with annotations
display(image)
```

```python
# Save the image with annotations
output_image_path = '/content/annotated_image.png'
image.save(output_image_path)

print(f"Annotated image saved to {output_image_path}")
```
